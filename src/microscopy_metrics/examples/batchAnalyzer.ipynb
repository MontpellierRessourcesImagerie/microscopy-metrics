### Exemple for the BatchAnalyzer module

First, you must enter the path to the folder containing all the images.

```python
import os

Folder = "/home/blangouet/Documents/perso/rapport/Presentation/test"
if not os.path.exists(Folder):
    raise ValueError(f"Folder {Folder} does not exist.")
```

Then, you must enter the detection and extraction parameters of the beads in the image.

```python
from microscopy_metrics.BatchAnalyzer import BatchAnalyzer

BatchAnalyzer = BatchAnalyzer(Folder)
BatchAnalyzer._detectionMethod = "Difference of Gaussian"
BatchAnalyzer._Sigma = 3.0
BatchAnalyzer._thresholdMethod = "legacy"
BatchAnalyzer._TheoreticalBeadSize = 0.2
BatchAnalyzer._ZRejectionMargin = 0.5
BatchAnalyzer._cropFactor = 10.0
BatchAnalyzer._thresholdIntensity = 0.5
BatchAnalyzer._pixelSize = [0.2, 0.1083, 0.1083]
```

Next, you need to define the parameters for calculating the metrics.

```python
BatchAnalyzer._MicroscopeType = "widefield"
BatchAnalyzer._numericalAperture = 1.4
BatchAnalyzer._emissionWavelength = 522.0
BatchAnalyzer._excitationWavelength = 488.0
BatchAnalyzer._refractionIndex = 1.518
```

We can now choose a fitting method.


```python
BatchAnalyzer._FitType = "2D rotation"
BatchAnalyzer._thresholdRSquared = 0.5
```

We finish by choosing the formats of image analysis reports.


```python
BatchAnalyzer._listReports = ["PDF", "HTML", "CSV"]
```

We can now start the batch analysis.

```python
BatchAnalyzer.analyze()
```
